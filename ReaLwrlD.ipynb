```python
import pandas as pd
# importing the pandas library usin the alias pd
```

```python
# looking at the Chicago public school data(CHICAGO_PUBLIC_SCHOOLS_DATA)

df = pd.read_csv("https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DB0201EN-SkillsNetwork/labs/FinalModule_Coursera_V5/data/ChicagoPublicSchools.csv")
zz = pd.DataFrame(df)
zz
```

```python
zz.columns
```

```python
# using the .to_sql() function to convert each csv file to a table in sqlite with the csv data loaded in it.


zz.to_sql("CHICAGO_PUBLIC_SCHOOLS_DATA", con, if_exists='replace', index=False, method="multi")
```

```python
# retrieving the list of table to see if the creation was successful

%sql SELECT name FROM sqlite_master WHERE type='table'

# in sqlite the system catalog table called sqlite_master contains the table metadata
```

```python
# looking at the schools table, how many columns are there ?

%sql SELECT count(name)  FROM PRAGMA_TABLE_INFO('CHICAGO_PUBLIC_SCHOOLS_DATA')

# the table seems to be correct
```

```python
# analysing the column data types

%sql SELECT name, type FROM PRAGMA_TABLE_INFO('CHICAGO_PUBLIC_SCHOOLS_DATA')
```

```python
# To know how many Elementary schools chicago has

%sql SELECT count(*) as 'No of ES' FROM CHICAGO_PUBLIC_SCHOOLS_DATA WHERE "Elementary, Middle, or High School"='ES'
```

```python
# to know how many Middle schools there are

%sql SELECT count(*) as 'No of MS' from CHICAGO_PUBLIC_SCHOOLS_DATA where "Elementary, Middle, or High School"='MS'
```

```python
# to know how many high schools there are

%sql SELECT count(*) as 'No of HS' from CHICAGO_PUBLIC_SCHOOLS_DATA where "Elementary, Middle, or High School"='HS'
```

```python
# looking at the schools with the highest safety score

%sql select Name_of_School, Safety_score, ("Elementary, Middle, or High School") as 'School' from CHICAGO_PUBLIC_SCHOOLS_DATA where Safety_score = (select MAX(Safety_score) from CHICAGO_PUBLIC_SCHOOLS_DATA)

# seeing that most of the schools are Elementary schools
```

```python
# Retrieving the school with the lowest safety score

%sql select Name_of_School, Safety_score from CHICAGO_PUBLIC_SCHOOLS_DATA where Safety_score = (select MIN(Safety_score) from CHICAGO_PUBLIC_SCHOOLS_DATA)
```

```python
# looking at the schools in ascending order of their safety scores

%sql select Name_of_School, Safety_score from CHICAGO_PUBLIC_SCHOOLS_DATA where Safety_score != 'None' order by Safety_score
```

```python
# looking at the top 20 schools with the Highest Average student attendance

%sql select Name_of_School, Average_Student_Attendance, ("Elementary, Middle, or High School") as 'School' from CHICAGO_PUBLIC_SCHOOLS_DATA order by Average_Student_Attendance desc nulls last limit 20
```

```python
# retrieving the list of 20 schools with the lowest average student attendance

%sql select Name_of_School, Average_Student_Attendance, ("Elementary, Middle, or High School") as 'School' from CHICAGO_PUBLIC_SCHOOLS_DATA order by Average_Student_Attendance limit 20
```

```python
# getting to know the average student attendance of the three types of schools

%sql select ROUND(AVG(Average_Student_Attendance), 2) as Average_att, ("Elementary, Middle, or High School") as 'School' from CHICAGO_PUBLIC_SCHOOLS_DATA GROUP BY "Elementary, Middle, or High School" order by Average_Student_Attendance DESC NULLS LAST
```

```python
# to get the total college enrollment for each community area with the total enrollment displayed in descending order

%sql select Community_Area_Name, SUM(College_Enrollment) as 'Total Enrollment' from CHICAGO_PUBLIC_SCHOOLS_DATA group by Community_Area_Name order by SUM(College_Enrollment) desc nulls last
```
